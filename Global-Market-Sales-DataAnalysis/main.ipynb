```python
import numpy as np
import pandas  as pd
import seaborn as sns
```

```python
data =pd.read_excel('Global Superstore Orders 2016.xlsx')
data.head()
```

```python
pd.crosstab(data.Segment,data.Category,values=data.Profit,aggfunc='mean')
```

```python
pd.crosstab(data.Segment,[data.Category,data.Market],values=data.Profit,aggfunc='mean')
```

```python

sns.heatmap(pd.crosstab([data.Market,data.Category],data.Segment),cmap='YlGnBu',annot=True,cbar=False)
```

```python
pd.crosstab(data.Market,data.Category,normalize=True)
```
